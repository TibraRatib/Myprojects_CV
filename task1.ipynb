<b>Task 1 : Image Transformation using OpenCV</b>

Using mr bean image code the following cells:

```python
#read image
import cv2
import numpy as np
import matplotlib.pyplot as plt
bean = cv2.imread("../bean.png")
```

```python
bean.shape
```

```python
plt.imshow(bean [:,:,::-1])
```

```python
beanHSV = cv2.cvtColor(bean, cv2.COLOR_BGR2HSV)
plt.imshow(beanHSV)
```

Q1: Image Translation -> shift of an image from one location to another

```python
shift_x, shift_y = 30, 50 
translation_matrix = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
translated_image = cv2.warpAffine(beanHSV, translation_matrix, (beanHSV.shape[1], beanHSV.shape[0]))
plt.imshow( translated_image)
```

```python
translated_image = cv2.warpAffine(bean, translation_matrix, (bean.shape[1], bean.shape[0]))
plt.imshow( translated_image [: ,: ,::-1])
```

Q2: Image Reflection -> flip the image horizontally & vertically

```python
# Flip the image vertically
ver_flip = cv2.flip(bean, 0)
plt.imshow(ver_flip [:,:,::-1])
```

```python
# Flip the image horizontally
hor_flip = cv2.flip(bean, 1)
plt.imshow(hor_flip[:,:,::-1])
```

Q3: Image Rotation -> rotate the image in any degree you want eg: (30, 60, 90)

```python
height, width = bean.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2),30, 0.6)
rotated_image = cv2.warpAffine(bean, rotation_matrix, (width, height))
plt.imshow(rotated_image[:,:,::-1])
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2),60, 0.6)
rotated_image = cv2.warpAffine(bean, rotation_matrix, (width, height))
plt.imshow(rotated_image[:,:,::-1])
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2),90, 0.6)
rotated_image = cv2.warpAffine(bean, rotation_matrix, (width, height))
plt.imshow(rotated_image[:,:,::-1])
```

Q4: Image Scaling -> resize image in any height and width you want eg: (480,640), (720, 800)

```python
resized1 = cv2.resize(bean, (480, 640))
plt.imshow(resized1[:,:,::-1])


```

```python
resized1 = cv2.resize(bean, (740, 800))
plt.imshow(resized1[:,:,::-1])
```

Q5: Image Cropping -> remove the background pixel of image (not importent)

```python
cropped_image = bean[50:300, 100:500]
plt.imshow(cropped_image[:,:,::-1])
```

```python
original_height, original_width = bean.shape[:2]
expanded_image = np.zeros((original_height, original_width, 3), dtype=np.uint8)
insert_region = expanded_image[50:300, 100:500]
insert_region[:, :] = cropped_image
plt.imshow(expanded_image[:,:,::-1])

```

```python
mask = np.zeros(bean.shape[:2], np.uint8)
rect = (50, 50, bean.shape[1] - 60, bean.shape[0] - 50)
cv2.grabCut(bean, mask, rect, None, None, 5, cv2.GC_INIT_WITH_RECT) # Apply GrabCut 
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
result = bean * mask2[:, :, np.newaxis]

plt.imshow(result[:,:,::-1])
```

Q6: Image Shearing -> shear the image once on X-Axis & Y-Axis

```python
#X-Axis
height, width = bean.shape[:2]
m = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(bean, m, (int(width*1.5), int(height*1.5)))
plt.imshow( sheared_img[:,:,::-1])
```

```python
#Y_Axis
m = np.float32([[1, 0, 0], [0.5, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(bean, m, (int(width*1.5), int(height*1.5)))
plt.imshow( sheared_img[:,:,::-1])
```
